```python
def fill_missing_months(data, month_col, value_col):
    """
    Fill missing months in a DataFrame with zero values.

    This function takes a DataFrame `data` containing monthly data, along with the column names for months (`month_col`)
    and corresponding values (`value_col`). It creates a DataFrame containing all twelve months, then merges it with the
    input DataFrame to ensure that all months are present. Missing values in the `value_col` column are filled with zeros.

    Parameters:
        data (DataFrame): The DataFrame containing the monthly data.
        month_col (str): The column name for the months.
        value_col (str): The column name for the corresponding values.

    Returns:
        DataFrame: The DataFrame with missing months filled, containing all twelve months with corresponding values.
    """
    all_months = ['January', 'February', 'March', 'April', 'May', 'June',
                  'July', 'August', 'September', 'October', 'November', 'December']

    full_months_df = pd.DataFrame(all_months, columns=[month_col])

    merged_df = pd.merge(full_months_df, data, on=month_col, how='left')

    merged_df[value_col].fillna(0, inplace=True)

    return merged_df
```

```python
def calculate_monthly_total_power(data, location, equipment, year, col_name):
    """
    Calculate the total power consumption on a monthly basis for a specific location, equipment, and year.

    This function takes in a DataFrame `data` containing power consumption data, and calculates the total power consumption
    on a monthly basis for the specified `location`, `equipment`, and `year`. The power consumption data should include a
    column specified by `col_name` containing timestamps.

    Parameters:
        data (DataFrame): The DataFrame containing power consumption data.
        location (str): Identifier for the location where the data was recorded.
        equipment (str): Identifier for the equipment being monitored.
        year (int): The year for which the data is being analyzed.
        col_name (str): The column name containing timestamps.

    Returns:
        DataFrame: DataFrame with monthly total power consumption data, with columns for month and total power consumed.
    """
    # Ensure the 'timestamps' column is in datetime format
    data[col_name] = pd.to_datetime(data[col_name])

    # Extract year and month from the datetime
    data['year'] = data[col_name].dt.year
    data['month'] = data[col_name].dt.month

    # Filter the data based on the input location, equipment, and year
    filtered_data = data[(data['location'] == location) & (data['equipment'] == equipment) & (data['year'] == year)]

    # Group by year and month, then calculate the sum of used power
    result = filtered_data.groupby(['month'])['used_power'].sum().reset_index()

    return result
# Example usage:
result = calculate_monthly_total_power(df, 'france', 'air_conditioning',2022, 'timestamps')
print(result)
```

```python
import calendar

def convert_numeric_month_to_name(df, column_name):
    # Map the numeric months to month names using the calendar module
    df[column_name] = df[column_name].apply(lambda x: calendar.month_name[x])

    return df
```

```python
def calculate_monthly_mean_temp(data, location, equipment, year, col_name):
    """
    Calculate the monthly mean temperature for a specific location and equipment in a given year.

    This function takes a DataFrame `data` containing temperature data with a timestamp column specified by `col_name`.
    It ensures that the timestamp column is in datetime format, extracts the year and month from the datetime,
    filters the data based on the input `location`, `equipment`, and `year`, and then calculates the mean temperature
    for each month.

    Parameters:
        data (DataFrame): DataFrame containing temperature data with a timestamp column.
        location (str): Identifier for the location where the temperature data was recorded.
        equipment (str): Identifier for the equipment for which temperature data is being analyzed.
        year (int): The year for which the temperature data is being analyzed.
        col_name (str): Name of the column containing timestamps.

    Returns:
        DataFrame: DataFrame containing the monthly mean temperature data for the specified location, equipment, and year.
    """
    # Ensure the 'timestamps' column is in datetime format
    data[col_name] = pd.to_datetime(data[col_name])

    # Extract year and month from the datetime
    data['year'] = data[col_name].dt.year
    data['month'] = data[col_name].dt.month

    # Filter the data based on the input location, equipment, and year
    filtered_data = data[(data['location'] == location) &
                         (data['equipment'] == equipment) &
                         (data['year'] == year)]

    # Group by year and month, then calculate the mean temperature
    result = filtered_data.groupby(['month'])['temperature'].mean().reset_index()

    return result

# Example usage:
result = calculate_monthly_mean_temp(df, 'france', 'air_conditioning',2022, 'timestamps')
print(result)
```

```python
def merge_dataframes(df1, df2, col_name):
    """
    Merge two dataframes based on a common column.

    Parameters:
        df1 (DataFrame): DataFrame containing power consumption data.
        df2 (DataFrame): DataFrame containing temperature data.
        col_name (str): Column name to merge the dataframes on. Default is 'month'.

    Returns:
        DataFrame: Merged DataFrame.
    """
    # Merge the two DataFrames on merge_column
    merged_df = pd.merge(df1, df2, on=col_name)

    return merged_df
```

```python
from matplotlib.ticker import FuncFormatter
from matplotlib.markers import MarkerStyle

def combined_plot_line_and_bar(df):
    """
    Plot power consumption and temperature for each month.

    Parameters:
        df (DataFrame): Merged DataFrame containing power consumption and temperature data.
    """
    # Create subplots with shared x-axis
    fig, ax1 = plt.subplots(figsize=(24, 8))

    # Plot bar chart for power consumption on the first subplot
    sns.barplot(x='month', y='used_power', data=df, ax=ax1, color='lightblue', label='Power Consumed', width=0.5)
    ax1.set_ylabel('Power Consumption', color='blue')

    # Create a second Y-axis for the line plot
    ax2 = ax1.twinx()
    sns.lineplot(x='month', y='temperature', data=df, ax=ax2, color='red', marker='D', label='Temperature',linestyle='dotted')
    ax2.set_ylabel('Mean Temperature', color='red')

    # Set Y-axis limits
    ax1.set_ylim([0, 4e6])
    ax2.set_ylim([0, 40])

    # Customize y-axis ticks for power consumption
    formatter = FuncFormatter(lambda x, _: '{:,.0f}k'.format(x/10000))  # Convert to k units
    ax1.yaxis.set_major_formatter(formatter)


    # Tilt the month names by 45 degrees
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

    #Adding Legend
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right')

    # Adding title and customizing axes
    plt.title('Power Consumption and Mean Temperature by Month')
    ax1.set_xlabel('Month')
    # Show the plot
    plt.show()
```

```python
def generate_scatter_plot(merged_df, x, y, hue, palette, legend):
    """
    Plot a scatter plot between Total Power Usage and Mean Temperature for all the 12 months.

    Parameters:
        merged_df (DataFrame): Merged DataFrame containing power consumption and temperature data.
        x (str): The column name for the x-axis (default is 'temperature').
        y (str): The column name for the y-axis (default is 'used_power').
        hue (str): The column name whose values will be used to color the points (default is 'month').
        palette (str or list of colors): Set of colors for mapping the hue variable (default is 'viridis').
        legend (str): How to draw the legend (default is 'full').
    """
    # Create a custom formatter function for y-axis ticks
    def y_axis_formatter(x, pos):
        if x!=0:
          return '{:,.0f}k'.format(x/10000)
        else:
          return 0

    # Create a scatter plot
    plt.figure(figsize=(8, 4))
    sns.scatterplot(x=x, y=y, data=merged_df, hue=hue, palette=palette, legend=legend)

    # Adjust y-axis ticks using the custom formatter
    plt.gca().yaxis.set_major_formatter(FuncFormatter(y_axis_formatter))

    # Adding title and labels
    plt.title('Scatter Plot: Mean Temperature vs. Total Power Usage')
    plt.xlabel('Mean Temperature')
    plt.ylabel('Total Power Usage (in 100k)')

    # Set Y-axis and X-axis limits
    plt.ylim(0, 3e6)
    plt.xlim(0, 35)

    # Show the plot
    plt.show()

# Example usage:
generate_scatter_plot(df_new,'temperature', 'used_power', 'month', 'plasma', 'full')
```

```python
def calculate_daily_total_power(data, location, equipment, year, col_name):
    """
    Calculate the daily total power consumption for a specific location, equipment, and year.

    This function takes a DataFrame `data` containing power consumption data and extracts daily total power consumption
    for the specified `location`, `equipment`, and `year` based on the provided column name `col_name` representing timestamps.

    Parameters:
        data (DataFrame): The DataFrame containing power consumption data.
        location (str): Identifier for the location where the data was recorded.
        equipment (str): Identifier for the equipment being monitored.
        year (int): The year for which the data is being analyzed.
        col_name (str): Column name representing timestamps.

    Returns:
        DataFrame: A DataFrame containing the daily total power consumption for the specified location, equipment, and year.
                   The DataFrame has columns for month, day, and the sum of power consumption.
    """
    # Ensure the timestamps column is in datetime format
    data[col_name] = pd.to_datetime(data[col_name])

    # Extract year, month, and day from the timestamps
    data['year'] = data[col_name].dt.year
    data['month'] = data[col_name].dt.month
    data['day'] = data[col_name].dt.day

    # Filter the data based on the input location, equipment, and year
    filtered_data = data[(data['location'] == location) & (data['equipment'] == equipment) & (data['year'] == year)]

    # Group by month and day, then calculate the sum of used power
    result = filtered_data.groupby(['month', 'day'])['used_power'].sum().reset_index()

    return result

# Example usage:
result = calculate_daily_total_power(df, 'france', 'air_conditioning',2022,'timestamps')

df2_power = result.copy()
```

```python
def calculate_daily_mean_temp(data, location, equipment, year, col_name):
    """
    Calculate the daily mean temperature for a specific location and equipment in a given year.

    This function takes a DataFrame `data` containing temperature data with a timestamp column specified by `col_name`.
    It first converts the timestamp column to datetime format and extracts the year, month, and day information.
    Then, it filters the data based on the provided `location`, `equipment`, and `year`.
    Next, it calculates the daily mean temperature by grouping the filtered data by month and day and computing the mean temperature.
    The result is returned as a DataFrame.

    Parameters:
        data (DataFrame): The DataFrame containing temperature data.
        location (str): Identifier for the location.
        equipment (str): Identifier for the equipment.
        year (int): The year for which the data is being analyzed.
        col_name (str): The name of the timestamp column.

    Returns:
        DataFrame: A DataFrame containing the daily mean temperature for the specified location, equipment, and year.
    """
    # Ensure the timestamp column is in datetime format
    data[col_name] = pd.to_datetime(data[col_name])

    # Extract year, month, and day from the datetime
    data['year'] = data[col_name].dt.year
    data['month'] = data[col_name].dt.month
    data['day'] = data[col_name].dt.day

    # Filter the data based on the input location, equipment, and year
    filtered_data = data[(data['location'] == location) & (data['equipment'] == equipment) & (data['year'] == year)]

    # Group by month and day, then calculate the mean temperature
    result = filtered_data.groupby(['month', 'day'])['temperature'].mean().reset_index()

    return result

# Example usage:
result = calculate_daily_mean_temp(df, 'france', 'air_conditioning',2022, 'timestamps')

df2_temp = result.copy()
```

```python
def plot_area_chart(df, x_col, y_col, xlabel, ylabel, title):
    """
    Generate an area chart using Seaborn and Matplotlib.

    Parameters:
        df (DataFrame): The DataFrame containing the data.
        x_col (str): The column name for the x-axis.
        y_col (str): The column name for the y-axis.
        xlabel (str): Label for the x-axis.
        ylabel (str): Label for the y-axis.
        title (str): Title for the plot.
    """
    plt.figure(figsize=(24, 8))
    sns.lineplot(x=x_col, y=y_col, data=df, color='skyblue', linewidth=0, alpha=0.6)
    plt.fill_between(df[x_col], df[y_col], color='skyblue', alpha=0.4)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

# Example usage:
plot_area_chart(newdf, 'serial', 'used_power', 'Days', 'Total Power Usage', 'Daily Total Power Usage')
```

```python
def plot_daywise_dual_axis_chart(df, x_col, y1_col, y2_col, xlabel, y1_label, y2_label, title):
    """
    Plot a combined line chart with dual y-axes for two different columns in a DataFrame.

    Parameters:
        df (DataFrame): The DataFrame containing the data.
        x_col (str): The column name for the x-axis.
        y1_col (str): The column name for the first y-axis.
        y2_col (str): The column name for the second y-axis.
        xlabel (str): The label for the x-axis.
        y1_label (str): The label for the first y-axis.
        y2_label (str): The label for the second y-axis.
        title (str): The title of the plot.
    """
    # Create a figure and an axis object
    fig, ax1 = plt.subplots(figsize=(24, 8))

    # Plot the area chart for power consumption on ax1
    sns.lineplot(data=df, x=x_col, y=y1_col, ax=ax1, color='#0021f3', linewidth=1, alpha=0.6, label=y1_label)
    ax1.fill_between(df[x_col], df[y1_col], color='#0021f3', alpha=0.4)
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(y1_label, color='black')
    ax1.tick_params(axis='y', labelcolor='black')


    # Create a second y-axis for the mean temperature line plot
    ax2 = ax1.twinx()
    sns.lineplot(data=df, x=x_col, y=y2_col, ax=ax2, linestyle='dotted', color='red', label=y2_label)
    ax2.set_ylabel(y2_label, color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    # Set Y-axis limits
    ax1.set_ylim([0, 200000])
    ax2.set_ylim([0, 30])

    # Customize y-axis ticks for power consumption

    formatter = FuncFormatter(lambda x, _: '{:,.0f}k'.format(x/1000))  # Convert to k units
    ax1.yaxis.set_major_formatter(formatter)

    #Adding Legend
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right')


    # Adding title and customizing axes
    plt.title('Total Power Usage and Mean Temperature by Day')
    ax1.set_xlabel('Days')
    ax1.set_ylabel('Total Power Usage')
    ax2.set_ylabel('Mean Temperature')

    # Title and show the plot
    plt.title(title)
    fig.tight_layout()  # Adjusts plot to fit titles
    plt.show()

# Example usage:
plot_daywise_dual_axis_chart(newdf, 'serial', 'used_power', 'temperature', 'Day', 'Total Power Usage', 'Mean Temperature', 'Daily Overview of Power Usage and Mean Temperature')
```

```python
def preprocess_data(data, col_name):
    """
    Preprocesses the data by converting the timestamp column to datetime format and extracting year, month, and day.

    Parameters:
        data (DataFrame): The DataFrame containing temperature data.
        col_name (str): The name of the timestamp column.

    Returns:
        DataFrame: Preprocessed DataFrame with year, month, and day columns added.
    """
    data[col_name] = pd.to_datetime(data[col_name])
    data['year'] = data[col_name].dt.year
    data['month'] = data[col_name].dt.month
    data['day'] = data[col_name].dt.day
    return data
```

```python
def filter_data(data, location, equipment, year):
    """
    Filters the data based on the provided location, equipment, and year.

    Parameters:
        data (DataFrame): The DataFrame containing temperature data.
        location (str): Identifier for the location.
        equipment (str): Identifier for the equipment.
        year (int): The year for which the data is being analyzed.

    Returns:
        DataFrame: Filtered DataFrame based on location, equipment, and year.
    """
    filtered_data = data[(data['location'] == location) &
                         (data['equipment'] == equipment) &
                         (data['year'] == year)]
    return filtered_data
```

```python
def calculate_daily_means(data):
    """
    Calculates the daily mean temperature for a DataFrame containing preprocessed temperature data.

    Parameters:
        data (DataFrame): Preprocessed DataFrame containing temperature data.

    Returns:
        DataFrame: A DataFrame containing the daily mean temperature.
    """
    result = data.groupby(['month', 'day'])['temperature'].mean().reset_index()
    return result
```

```python
def calculate_hourly_total_power(data, location, equipment, year, month, day, col_name):
    """
    Calculate the hourly total power usage for a single day of a particular month.

    Parameters:
        data (DataFrame): The DataFrame containing power consumption data.
        location (str): Identifier for the location where the data was recorded.
        equipment (str): Identifier for the equipment being monitored.
        year (int): The year for which the data is being analyzed.
        month (int): The month for which the data is being analyzed.
        day (int): The day for which the hourly total power usage will be calculated.
        col_name (str): The column name containing timestamps.

    Returns:
        DataFrame: DataFrame with hourly total power usage data for the specified day.
    """
    # Ensure the 'timestamps' column is in datetime format
    data[col_name] = pd.to_datetime(data[col_name])

    # Extract year, month, and day from the datetime
    data['year'] = data[col_name].dt.year
    data['month'] = data[col_name].dt.month
    data['day'] = data[col_name].dt.day
    data['hour'] = data[col_name].dt.hour

    # Filter the data based on the input location, equipment, year, month, and day
    filtered_data = data[(data['location'] == location) &
                         (data['equipment'] == equipment) &
                         (data['year'] == year) &
                         (data['month'] == month) &
                         (data['day'] == day)]

    # Group by hour and calculate the sum of used power
    result = filtered_data.groupby(['hour'])['used_power'].sum().reset_index()

    return result

# Example usage:
result = calculate_hourly_total_power(df, 'france', 'air_conditioning', 2022, 4, 27, 'timestamps')
df3_power = result.copy()
print(result)
```

```python
def calculate_hourly_mean_temperature(data, location, equipment, year, month, day, col_name):
    """
    Calculate the hourly mean temperature for a single day of a particular month.

    Parameters:
        data (DataFrame): The DataFrame containing temperature data.
        location (str): Identifier for the location where the data was recorded.
        equipment (str): Identifier for the equipment being monitored.
        year (int): The year for which the data is being analyzed.
        month (int): The month for which the data is being analyzed.
        day (int): The day for which the hourly mean temperature will be calculated.
        col_name (str): The column name containing timestamps.

    Returns:
        DataFrame: DataFrame with hourly mean temperature data for the specified day.
    """
    # Ensure the 'timestamps' column is in datetime format
    data[col_name] = pd.to_datetime(data[col_name])

    # Extract year, month, and day from the datetime
    data['year'] = data[col_name].dt.year
    data['month'] = data[col_name].dt.month
    data['day'] = data[col_name].dt.day
    data['hour'] = data[col_name].dt.hour

    # Filter the data based on the input location, equipment, year, month, and day
    filtered_data = data[(data['location'] == location) &
                         (data['equipment'] == equipment) &
                         (data['year'] == year) &
                         (data['month'] == month) &
                         (data['day'] == day)]

    # Group by hour and calculate the mean temperature
    result = filtered_data.groupby(['hour'])['temperature'].mean().reset_index()

    return result

# Example usage:
result = calculate_hourly_mean_temperature(df, 'france', 'air_conditioning', 2022, 4, 27, 'timestamps')
df3_temp = result.copy()
print(result)
```

```python
def plot_hourly_power_and_temp(df, x_col, y1_col, y2_col, xlabel, y1_label, y2_label, title):
    """
    Plot hourly power consumption and mean temperature on the same plot.

    Parameters:
        df (DataFrame): Merged DataFrame containing hourly power consumption and mean temperature data.
        x_col (str): The column name for the x-axis (hour).
        y1_col (str): The column name for the y-axis of the first line plot (power consumption).
        y2_col (str): The column name for the y-axis of the second line plot (mean temperature).
        xlabel (str): Label for the x-axis.
        y1_label (str): Label for the y-axis of the first line plot (power consumption).
        y2_label (str): Label for the y-axis of the second line plot (mean temperature).
        title (str): Title for the plot.
    """
    # Create subplots with shared x-axis
    fig, ax1 = plt.subplots(figsize=(24, 8))

    # Plot line chart for power consumption on ax1
    sns.lineplot(data=df, x=x_col, y=y1_col, ax=ax1, linestyle='dotted', color='blue', label=y1_label)

    # Set labels and title for ax1
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(y1_label, color='blue')

    # Create a second y-axis for the mean temperature line plot
    ax2 = ax1.twinx()
    sns.lineplot(data=df, x=x_col, y=y2_col, ax=ax2, linestyle='dotted',color='red', label=y2_label)

    # Set labels for ax2
    ax2.set_ylabel(y2_label, color='red')

    # Set ticks and labels for x-axis (hours)
    ax1.set_xticks(range(0, 21, 4))
    ax1.set_xticklabels([str(i) for i in range(0, 21, 4)])

    # Set dynamic scale for y-axis (power consumption in kW)
    # max_power = df[y1_col].max()
    ax1.set_ylim(0, 15000)
    ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f} kW'.format(x/1000)))

    # Set dynamic scale for y-axis (mean temperature)
    # max_temp = df[y2_col].max()
    ax2.set_ylim(0, 20)

    # Adding Legend
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right')

    # Adding title
    plt.title(title)

    # Show the plot
    plt.show()

# Example usage:
plot_hourly_power_and_temp(final_df, 'hour', 'used_power', 'temperature', 'Hours', 'Power Consumption(kWh)', 'Mean Temperature(°C)', 'Hourly Overview of Power Consumption and Mean Temperature')
```

```python
import ipywidgets as widgets
from ipywidgets import interact

def plot_power_consumption_scale(df, location, year, month, day, equipment, hour):
    """
    Plot a color scale indicating power consumption for a specific location, year, month, day, and hour.

    Parameters:
        df (DataFrame): DataFrame containing power consumption data.
        location (str): Location identifier.
        year (int): Year.
        month (int): Month.
        day (int): Day.
        equipment (str): Equipment identifier.
        hour (int): Hour of the day.
    """
    # Filter the DataFrame for the selected location, year, month, day, and equipment
    df_filtered = df[(df['location'] == location) & (df['year'] == year) & (df['month'] == month) & (df['day'] == day) & (df['equipment'] == equipment)]

    # Calculate the maximum power consumption for the selected day
    max_power_consumption = df_filtered['used_power'].max()

    # Convert max_power_consumption to kilowatts
    max_power_consumption_kw = max_power_consumption / 1000  # Convert to kW

    # Create the color scale
    plt.figure(figsize=(6, 2))
    cmap = sns.color_palette("inferno", as_cmap=True)
    cmap_reversed = cmap.reversed()
    norm = plt.Normalize(0, max_power_consumption_kw)  # Normalize the color scale
    plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap_reversed), orientation='horizontal', ticks=[0, max_power_consumption_kw])
    plt.title(f'Power Consumption Scale for {location}, Year: {year}, Month: {month}, Day: {day}, Equipment: {equipment}')
    plt.xlabel('Power Consumption (kW)')

    # Set the x-limits to keep the scale fixed
    plt.xlim(0, max_power_consumption_kw)

    # Plot marker indicating power consumption for the selected hour
    if hour is not None:
        power_consumption_hour = df_filtered[df_filtered['hour'] == hour]['used_power'].values
        power_consumption_hour_kw = power_consumption_hour / 1000  # Convert to kW
        plt.axvline(x=power_consumption_hour_kw, color='red', linestyle='--', label=f'Hour {hour} Power Usage: {power_consumption_hour_kw} kW')
        plt.legend()

    # Show the color scale
    plt.show()

# Create dropdowns for selecting location, year, month, day, and equipment
location_dropdown = widgets.Dropdown(options=df['location'].unique(), description='Location:')
year_dropdown = widgets.Dropdown(options=df['year'].unique(), description='Year:')
month_dropdown = widgets.Dropdown(options=df['month'].unique(), description='Month:')
day_dropdown = widgets.Dropdown(options=df['day'].unique(), description='Day:')
equipment_dropdown = widgets.Dropdown(options=df['equipment'].unique(), description='Equipment:')
hour_slider = widgets.IntSlider(value=0, min=0, max=23, step=1, description='Hour:')

# Define the interaction between the widgets and the plot
interact(plot_power_consumption_scale,
         df=widgets.fixed(df),
         location=location_dropdown,
         year=year_dropdown,
         month=month_dropdown,
         day=day_dropdown,
         equipment=equipment_dropdown,
         hour=hour_slider)
```
