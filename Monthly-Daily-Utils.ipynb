```python
def plot_using_seaborn(data, x_col, y_col, x_lab, y_lab, title):
  sns.barplot(x='Month', y='Power Usage (kWh)', data=data)

  plt.xlabel(x_lab)
  plt.ylabel(y_lab)
  plt.title(title)

  plt.xticks(rotation=45)
  plt.show()
```

```python
def plot_seaborn(data, x_col, y_col,x_lab,y_lab,title,size=(10, 6), color='blue'):
    plt.figure(figsize=size)
    sns.barplot(x=x_col, y=y_col, data=data, color=color)

    plt.xlabel(x_lab)
    plt.ylabel(y_lab)
    plt.title(title)

    plt.xticks(rotation=45)

    plt.show()
```

```python
plot_seaborn(data = data, x_col = 'Month', y_col = 'Power Usage (kWh)',
             x_lab = 'Months', y_lab = 'Monthly Power Usage',
             title = 'Monthly Power Usage per Month',
             size = (10, 8),
             color = 'red')
```

```python
def create_hv_bar_chart(data, x_col, y_col, xlabel, ylabel, title):

    hv.extension('bokeh')
    bar_chart = hv.Bars(data, [x_col], [y_col])

    bar_chart = bar_chart.opts(
        opts.Bars(
            width=800, height=400, tools=['hover'],
            xlabel=xlabel, ylabel=ylabel, title=title,
            color='blue', show_grid = True, invert_axes=False
        )
    )

    return bar_chart
```

```python
create_hv_bar_chart(data, x_col = 'Month', y_col = 'Power Usage (kWh)', xlabel='Months', ylabel='Monthly Power Usage', title='Monthly Power Usage per Month')
```

```python
def plot_holoviews(data, x_col, y_col, xlabel, ylabel, title, size = (800,400),color='blue'):

    hv.extension('bokeh')
    bar_chart = hv.Bars(data, [x_col], [y_col])

    bar_chart = bar_chart.opts(
        opts.Bars(
            width=size[0], height=size[1], tools=['hover'],
            xlabel=xlabel, ylabel=ylabel, title=title,
            color=color, show_grid = True, invert_axes=False
        )
    )

    return bar_chart
```

```python
plot_holoviews(data, x_col = 'Month', y_col = 'Power Usage (kWh)',
               xlabel='Months', ylabel='Monthly Power Usage',
               title='Monthly Power Usage per Month',
               size = (800,600),
               color = 'red')
```

```python
def plot_power_usage(data, x_col, y_col, title, labels_dict, color_col, barmode='group', xaxis_tickangle=-45):

    fig = px.bar(data, x=x_col, y=y_col, title=title, labels=labels_dict,
                 color=color_col, barmode=barmode)

    # Update layout for better readability
    fig.update_layout(xaxis_tickangle=xaxis_tickangle)

    # Display the plot
    fig.show()
```

```python
def plot_ploty(data, x_col, y_col, title, labels_dict, color_col, barmode='group', xaxis_tickangle=-45, size = (800, 600)):

    fig = px.bar(data, x=x_col, y=y_col, title=title, labels=labels_dict,
                 color=color_col, barmode=barmode)

    fig.update_layout(xaxis_tickangle=xaxis_tickangle, width=size[0], height=size[1])

    fig.show()
```

```python
def fill_missing_months(data, month_col, value_col):
    all_months = ['January', 'February', 'March', 'April', 'May', 'June',
                  'July', 'August', 'September', 'October', 'November', 'December']

    full_months_df = pd.DataFrame(all_months, columns=[month_col])

    merged_df = pd.merge(full_months_df, data, on=month_col, how='left')

    merged_df[value_col].fillna(0, inplace=True)

    return merged_df
```

# **TimeStamp**


```python
def calculate_monthly_mean_power(data, location, equipment, year):
    # Ensure the 'timestamps' column is in datetime format
    data['timestamps'] = pd.to_datetime(data['timestamps'])

    # Extract year and month from the datetime
    data['year'] = data['timestamps'].dt.year
    data['month'] = data['timestamps'].dt.month

    # Filter the data based on the input location and equipment
    filtered_data = data[(data['location'] == location) & (data['equipment'] == equipment) & (data['year'] == year)]

    # Group by year and month, then calculate the mean used power
    result = filtered_data.groupby(['year', 'month'])['used_power'].sum().reset_index()

    return result

# Example usage:
result = calculate_monthly_mean_power(df, 'france', 'air_conditioning',2022)
print(result)
```

```python
result = result.drop(columns=['year'])
```

```python
import calendar

def convert_numeric_month_to_name(df, column_name):
    # Map the numeric months to month names using the calendar module
    df[column_name] = df[column_name].apply(lambda x: calendar.month_name[x])

    return df
```
